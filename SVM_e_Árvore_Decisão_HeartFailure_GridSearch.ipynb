```python
import numpy as np
from sklearn import model_selection
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import pandas as pd
from sklearn.model_selection import GridSearchCV
from sklearn import preprocessing

#Carregando dados do csv
data = pd.read_csv('heart_failure_clinical_records_dataset.csv')

#Segregando valores e nomes de colunas 
dta = data.values
att = data.columns

#Separação entre X e y
X=dta[:,0:12]
y=dta[:,12]

#Normalização
#scaler = preprocessing.StandardScaler().fit(X)
#X = scaler.transform(X)

scaler = preprocessing.MinMaxScaler().fit(X)
X = scaler.transform(X)

#X = preprocessing.normalize(X, norm='l2', axis=0)

parameters = [
      {'C':[1,5,10,50,100,200],'kernel':['linear']},
      {'C':[1,5,10,50,100,200],'kernel':['poly']},
      {'C':[1,5,10,50,100,200],'kernel':['rbf'], 'gamma':[0.1,0.01,0.001,'scale']},
]

parameters_tree = [
      {'criterion':['gini', 'entropy'],
       'max_depth':[3,5,10],
       'min_samples_split':[2,5,10],
       'min_samples_leaf':[1,2,5,10]}
]

#Criação do indutor
folds=10
#clf = SVC(probability=True)
clf = DecisionTreeClassifier()
gs = GridSearchCV(clf, parameters_tree, scoring='accuracy', cv=folds, n_jobs=5)
gs.fit(X,y)
clf = gs.best_estimator_

#Apresentação resultados GridSearch
from tabulate import tabulate
df=gs.cv_results_
print(tabulate(df, headers='keys', tablefmt='psql'))

print("Melhores Parâmetros")
print(gs.best_params_)

#Utilização de validação cruzada com melhores atributos do indutor
result=model_selection.cross_val_score(clf, X, y, cv=folds)

#Impressão dos resultados de média e desvio padrão
print("\nCross Validation Results %d folds: " % folds)
print("Mean Accuracy: %.5f" % result.mean())
print("Mean STD: %.5f" % result.std())


#Realizando predições e mostrando a matriz de confusão
predicted = model_selection.cross_val_predict(clf, X, y, cv=folds)
matriz = confusion_matrix(y, predicted)
print("\nConfusion Matrix: ")
print(matriz)
```
